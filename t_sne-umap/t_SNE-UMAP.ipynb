```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import plotly
```

### Generate synthetic data with four clusters in a 3D space


```python
# cluster centers initialization

cluster_centers = [[2, -6, -6], [-1, 9, 4], [-8, 7, 2], [4, 7, 9]]

# Cluster standard deviations:
cluster_std=[1,1,2,3.5]
```

```python
# Make the blobs and return the data and the blob labels
from sklearn.datasets import make_blobs

x,labels_ = make_blobs(n_samples= 500, n_features=3, cluster_std=cluster_std, centers=cluster_centers, random_state=42)
```

### Display the data in an interactive Plotly 3D scatter plot


```python
import plotly.express as px
# Create a DataFrame for Plotly
df =pd.DataFrame(x, columns=['X','Y','Z'])

#create intercatice 3Dscatter plot
fig = px.scatter_3d(df, x = 'X', y = 'Y', z = 'Z', color=labels_.astype(str),
                    opacity=0.7, color_discrete_sequence=px.colors.qualitative.G10, 
                    title="3D Scatter Plot of Four Blobs")

fig.update_traces(marker=dict(size=5, line=dict(width=1, color='black')), showlegend=False)
fig.update_layout(coloraxis_showscale=False, width=1000, height=800)  # Remove color bar, resize plot

fig.show()
```
